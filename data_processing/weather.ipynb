# Weather Data

```python
weather_df = pd.read_csv('data/weather.csv')
```

```python
weather_df.head()
```

```python
len(weather_df.state.unique())
```

```python
weather_df = weather_df[['month', 'state', 'average_temp']]
weather_df = weather_df.groupby(['month', 'state']).mean().reset_index()
weather_df
```

```python
# Rename columns
weather_df.columns = ['month', 'state', 'temperature']
```

```python
# Round temperature
weather_df.temperature = weather_df.temperature.round(2)
len(weather_df)
```

```python
# Join to get state id
state = pd.read_csv("data/state.csv", dtype={
    'state_id': 'string',
})
weather_df = weather_df.merge(right=state, left_on='state', right_on='name', how='inner')
weather_df = weather_df[['state_id', 'state', 'month', 'temperature']]
weather_df
```

```python
weather_df['month'] = weather_df['month'].astype(int)
weather_df.to_csv('data/processed_weather.csv', index=False)
weather_df
```
